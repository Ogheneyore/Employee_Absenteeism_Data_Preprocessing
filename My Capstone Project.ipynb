### Dummy Encoding
Dummy variables are a placeholder for categorical variables.

I want to encode the nominal variable Reason for Absence numerically, but without implying some type of ordering.

I had to convert the reason for absence column entries into 0s and 1s. The way to do this is by creating a number of columns corresponding to the number of unique reasons for absence. I have to then set each observation to 1 for each column to which the reasons correspond.

I created a dummy variables matrix encoding these reasons using pandas:

```python
#700 observations vs 28 reasons for absence.
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first=True)
reason_columns
```

```python
df.columns.values
```

```python
df = df.drop(['Reason for Absence'], axis=1)
```

```python
df
```

### Checking out Reasons for Staff Absence
This column contains information on the staff’ self-reported reasons for non-attendance.

```python
#reason 1-14 = various diseases

#reason 15-17 = pregnancies

#reason 18-21 = poisoning-related

#reasons 22-28 = light reasons

reason_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_4 = reason_columns.loc[:, 22:].max(axis=1)
```

```python
 # Shows that the 4 reasons columns have been appended to the end of the dataframe.
df = pd.concat([df, reason_1, reason_2, reason_3, reason_4], axis=1)
df
```

```python
df.columns.values
```

```python
column_names=['Date', 'Transportation Expense', 'Distance to Work', 'Age',
              'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
              'Pets', 'Absenteeism Time in Hours', 'reason_1',' reason_2', 'reason_3', 'reason_4']
```

```python
df.columns=column_names
df
```

## Using Checkpoint

```python
def checkpoint(file_name,checkpoint_header,checkpoint_data):
    
    np.savez(file_name,header = checkpoint_header,data = checkpoint_data)
    
    checkpoint_variable = np.load(file_name + ".npz")
    return(checkpoint_variable)
```

```python
checkpoint_1=checkpoint('checkpoint_1',column_names,df)
```

```python
checkpoint_1['header']
```
